```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn import datasets
```

```python
iris = datasets.load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target']=iris.target
```

```python
# independent variable
```

```python
x = iris_df.drop('target', axis=1)
x
```

```python
#dependent variable
```

```python
y= iris_df['target']
y
```

```python
# train test split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
#model creation
```

```python
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(x_train,y_train)
```

```python
#prediction
```

```python
y_pred=knn.predict(x_test)
```

```python
y_pred
```

```python
x_pred=knn.predict(x_train)
x_pred
```

```python
acc = accuracy_score(y_test,y_pred)
acc
```

```python
#visualization of testing data
```

```python
plt.scatter(x_test.iloc[:,0],x_test.iloc[:,1],cmap='viridis',c=y_pred,edgecolor='red',s=50)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("K Neighbor Classifier")
plt.show()
```

```python
#visualizing training data
```

```python
plt.scatter(x_train.iloc[:,2],x_train.iloc[:,3],cmap='viridis',c=x_pred,edgecolor='red',s=50)
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("KNeighbors Classifier")
plt.show()
```

```python
# visualizing testing data
```

```python
plt.scatter(x_test.iloc[:,2],x_test.iloc[:,3],cmap='viridis',c=y_pred,edgecolor='red',s=50)
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("KNeighbors Classifier")
plt.show()
```

```python
new_data_point = [[5.5, 3.2,4.5,1.5]]
```

```python
# Predict the class of the new data point
new_data_pred = knn.predict(new_data_point)
print(new_data_pred)
```

```python
# Visualize the new data point
plt.scatter(x_test.iloc[:, 0], x_test.iloc[:, 1], c=y_pred, cmap='viridis', edgecolor='red', s=50, label='Predicted')
plt.scatter(x_train.iloc[:, 0], x_train.iloc[:, 1], c=y_train, cmap='viridis', edgecolor='black', s=50, alpha=0.5, label='Train')
plt.scatter(new_data_point[0][0], new_data_point[0][1], color='blue', marker='*', s=200, label='New Data Point')
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("K Nearest Neighbors Classifier with New Data Point")
plt.show()
```

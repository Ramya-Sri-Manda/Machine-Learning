```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\ramya\\Downloads\\yield.csv")
df
```

```python
y=df['Value']
y
```

```python
x=df.drop('Value',axis=1)
x
```

```python
df.dropna(inplace=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder
lb= LabelEncoder()
df['Domain Code']=lb.fit_transform(df['Domain Code'])
df['Domain']=lb.fit_transform(df['Domain'])
df['Area']=lb.fit_transform(df['Area'])
df['Element']=lb.fit_transform(df['Element'])
df['Item']=lb.fit_transform(df['Item'])
df['Unit']=lb.fit_transform(df['Unit'])
```

```python
df
```

```python
x=df.drop('Value',axis=1)
x
```

```python
y=df['Value']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor(max_depth=2,random_state=100)
```

```python
rf.fit(x_train,y_train)
```

```python
#making prediction
```

```python
y_rf_train_pred=rf.predict(x_train)
y_rf_test_pred=rf.predict(x_test)
```

```python
#evaluating
```

```python
from sklearn.metrics import mean_squared_error,r2_score,accuracy_score
```

```python
rf_train_mse=mean_squared_error(y_train,y_rf_train_pred)
rf_train_r2=r2_score(y_train,y_rf_train_pred)
```

```python
rf_test_mse=mean_squared_error(y_test,y_rf_test_pred)
rf_test_r2=r2_score(y_test,y_rf_test_pred)
```

```python
rf_results=pd.DataFrame(['RandomForest',rf_train_mse,rf_train_r2,rf_test_mse,rf_test_r2]).transpose()
rf_results.columns=['Method','Training MSE','Training R2','Test MSE','Test R2']
rf_results
```

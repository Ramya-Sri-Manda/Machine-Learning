```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score
import pickle  # Import pickle module

```

```python
crop = pd.read_csv("Crop_recommendation.csv")
crop
```

```python
x = crop.drop('label', axis=1)
y = crop['label']
```

```python
crop.dropna(inplace=True)
crop
```

```python
lb = LabelEncoder()
crop['label'] = lb.fit_transform(crop['label'])
crop

```

```python
x = crop.drop('label', axis=1)
y = crop['label']

```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.8, random_state=0)

```

```python
classifier = RandomForestClassifier(n_estimators=10, criterion="entropy")
classifier.fit(x_train, y_train)
```

```python
x_pred = classifier.predict(x_train)
y_pred = classifier.predict(x_test)
```

```python
mean = mean_squared_error(y_train, x_pred)
r2 = r2_score(y_train, x_pred) * 100
print("Train Mean Squared Error:", mean)
print("Train R^2 Score:", r2)
```

```python
mean = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Test Mean Squared Error:", mean)
print("Test R^2 Score:", r2)
```

```python
pred = classifier.predict(x_test)
ac = accuracy_score(y_test, y_pred)
print("Test Accuracy:", ac * 100)
```
